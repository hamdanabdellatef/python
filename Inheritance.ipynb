```python
class Automobile():
    def __init__(self, make, model, mileage, price):
        self.__make = make
        self.__model = model
        self.__mileage = mileage
        self.__price = price

    def set_make(self, make):
        self.__make = make

    def set_model(self, model):
        self.__model = model

    def set_mileage(self, mileage):
        self.__mileage = mileage

    def set_price(self, price):
        self.__price = price

    def get_make(self):
        return self.__make

    def get_model(self):
        return self.__model

    def get_mileage(self):
        return self.__mileage

    def get_price(self):
        return self.__price
```

```python
BMW = Automobile('BMW', 2001, 70000, 15000.0)
```

```python
print(BMW.get_make())
print(BMW.get_model())
print(BMW.get_mileage())
print(BMW.get_price())
```

```python
BMW.set_price(5000.0)
print(BMW.get_price())
```

```python
class Car(Automobile):
    def __init__(self, make, model, mileage, price, doors):
        Automobile.__init__(self, make, model, mileage, price)
        self.__doors = doors

    def set_doors(self, doors):
        self.__doors = doors

    def get_doors(self):
        return self.__doors

    def get_price(self):
        if self.__doors == 2:
            return Automobile.get_price(self) * 1.25
        elif self.__doors == 4:
            return Automobile.get_price(self)
```

```python
CarA = Car('BMW', 2001, 70000, 5000.0, 4)
CarB = Car('Audi',2024, 0, 100000, 2)
```

```python
print(isinstance(CarA,Car))
print(isinstance(CarA,Automobile))
```

```python
print(CarA.get_price())
```

```python
print(CarB.get_price())
```

```python
import vehicles

car = vehicles.Car('BMW', 2001, 70000, 15000.0, 4)
truck = vehicles.Truck('Toyota', 2002, 40000, 12000.0, '4WD')
suv = vehicles.SUV('Volvo', 2000, 30000, 18500.0, 5)

print('USED CAR INVENTORY')
print('===================')

print('The following car is in inventory:')
print('Make:', car.get_make())
print('Model:', car.get_model())
print('Mileage:', car.get_mileage())
print('Price:', car.get_price())
print('Number of doors:', car.get_doors())
print()

# Display the truck's data.
print('The following pickup truck is in inventory.')
print('Make:', truck.get_make())
print('Model:', truck.get_model())
print('Mileage:', truck.get_mileage())
print('Price:', truck.get_price())
print('Drive type:', truck.get_drive_type())
print()

# Display the SUV's data.
print('The following SUV is in inventory.')
print('Make:', suv.get_make())
print('Model:', suv.get_model())
print('Mileage:', suv.get_mileage())
print('Price:', suv.get_price())
print('Passenger Capacity:', suv.get_pass_cap())
```

```python
def display_vehicle_info(vehicle):
    print('Make:', vehicle.get_make())
    print('Model:', vehicle.get_model())
    print('Mileage:', vehicle.get_mileage())
    print('Price:', vehicle.get_price())
    if isinstance(vehicle, vehicles.Car):
        print('Number of doors:', vehicle.get_doors())
    elif isinstance(vehicle, vehicles.Truck):
        print('Drive type:', vehicle.get_drive_type())
    elif isinstance(vehicle, vehicles.SUV):
        print('Passenger capacity:', vehicle.get_pass_cap())
```

```python
display_vehicle_info(truck)
```
