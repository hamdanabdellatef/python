### Example:
Random Number Frequencies

Write a program that generates 100 random numbers between 1 and 10. The program should store the frequency of each number generated in a dictionary with the number as the key and the amount of times it has occurred as the value. For example, if the program generates the number 6 a total of 11 times, the dictionary will contain a key of 6 with an associated value of 11. Once all of the numbers have been generated, display information about the frequency of each number.


````python
import random
rnd = []
for i in range(0,100):
    rnd.append(random.randint(1,10))

print(rnd)
````

freq = {1:how many 1s we have, 2: how many 2s we have, ..., 10: value}

````python
freq = {}

for x in rnd:
    if x not in freq:
        freq[x] = 1
    else:
        freq[x]+=1
print(freq)
````

````python
freq_orderd = {}
for key in range(1,11):
    freq_orderd[key] = rnd.count(key)
print(freq_orderd)
````

## set operations

Union: `A | B`

Intersection: `A & B`

Difference: `A - B`

Symmetric Difference: `A ^ B`

Subset: `A <= B`

Superset: `A >= B`

````python
set3 = set('Khalil is a top student')
print(set3)
````

````python
for item in set3:
    print(item)
````

````python
print('K' in set3)
````

````python
set1 = set('We use khalil in all of our examples')
set2 = set('Khalil is also in this example')
````

````python
print(set1)
print(set2)
````

````python
set3 = set1.union(set2)
set4 = set1 | set2

print(set3)
print(set4)
````

````python
set3 = set1.intersection(set2)
set4 = set1 & set2
print(set3)
print(set4)
````

````python
set3 = set1.difference(set2)
set4 = set1 - set2
print(set3)
print(set4)
````

````python
set3 = set1.symmetric_difference(set2)
set4 = set1 ^ set2
print(set3)
print(set4)
````

````python
print(set1.issubset(set2))
````

````python
setA = set('example')
print(setA.issubset(set1))
````

````python
print(set1.issuperset(setA))
````

````python
print(set1.issuperset(set2))
````

````python
import string
def read_words_from_file(filename):
    words = []
    with open(filename, 'r', encoding='utf-8') as file:
        for line in file:
            line = line.translate(str.maketrans('', '', string.punctuation))
            words.extend(line.split())
    return words
````
